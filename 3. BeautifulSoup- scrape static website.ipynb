```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://www.gov.uk/search/services')
```

```python
# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
s = soup.find('select', id="level_one_taxon")
content = s.find_all('option',{'data-track-action':"level_one_taxon"})
```

```python
level_one_taxon = {}
for c in content:
    level_one_taxon[c['data-track-label']] = c['value']
```

```python
level_one_taxon
```

```python
s = soup.find('select', id="level_two_taxon")
content = s.find_all('option',{'data-track-action':"level_two_taxon"})
```

```python
level_two_taxon = {}

for c in content:
    parent = c['data-topic-parent']
    if parent in level_two_taxon.keys():
        level_two_taxon[parent].append((c['data-track-label'],c['value']))
    else:
        level_two_taxon[parent] = [(c['data-track-label'],c['value'])]
```

```python
level_two_taxon
```

```python
r = requests.get('https://www.gov.uk/search/services?organisations%5B%5D=academy-for-social-justice&order=most-viewed')
soup = BeautifulSoup(r.content, 'html.parser')
s = soup.find('ul', class_="govuk-checkboxes gem-c-checkboxes__list")
content = s.find_all('input',{'class':"govuk-checkboxes__input"})
organisations = {}
for c in content:
    organisations[c['data-track-label']] = (c['name'],c['value'])
```
